```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, Birch
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_samples
df = pd.read_csv(r"C:\Users\DELL\OneDrive\Desktop\wine-clustering.csv")

print(df)
## PREPROCESS:
print(df.isna().sum())
class Plot_cluster_all:
    
    def __init__(self, data):
        self.data = data
    
    def remove_outliers_all(self):
        Q1 = self.data.quantile(0.25)
        Q3 = self.data.quantile(0.75)
        IQR = Q3 - Q1
        self.data = self.data[~((self.data < (Q1 - 1.5 * IQR)) | (self.data > (Q3 + 1.5 * IQR))).any(axis=1)]
        
    def plot_histogram_2(self):
        
        list_cols = self.data.columns.tolist()
        c1,c2 = np.random.choice(list_cols, 2, replace=False)
        plt.figure(figsize=(12, 6))  
        plt.subplot(1, 2, 1)
        sns.histplot(self.data[c1], kde=True)
        plt.title(f"Histogram of {c2}")
        
        plt.subplot(1, 2, 2)
        sns.histplot(self.data[c2], kde=True)
        plt.title(f"Histogram of {c2}")
        
        plt.tight_layout()
        plt.show()
    
    def plot_scatter_2(self):
        list_cols = self.data.columns.tolist()
        c1,c2 = np.random.choice(list_cols, 2, replace=False)
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x=self.data[c1], y=self.data[c2])
        plt.title(f"Scatter plot of {c1} vs {c2}")
        plt.xlabel(c2)
        plt.ylabel(c1)
        plt.show()
    
    def plot_pairplot_all(self):
        plt.figure(figsize=(10, 8))
        sns.pairplot(self.data)
        plt.show()
    
    def plot_heatmap_all(self):
        plt.figure(figsize=(10, 8))
        corr_matrix = self.data.corr()
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
        plt.title("Correlation Heatmap")
        plt.show()
    
    def plot_boxplot_all(self):
        plt.figure(figsize=(12, 8))
        sns.boxplot(data=self.data)
        plt.title("Boxplot for each feature")
        plt.xticks(rotation=45)
        plt.show()

    def visualize_all(self):
        self.remove_outliers_all()
        self.plot_histogram_2()
        self.plot_scatter_2()
        self.plot_pairplot_all()
        self.plot_heatmap_all()
        self.plot_boxplot_all()
visualize = Plot_cluster_all(df)
visualize.visualize_all()

# Class for KMeans Clustering
class KMeansClustering:
    def __init__(self, n_clusters=3):
        self.n_clusters = n_clusters
        self.model = KMeans(n_clusters=self.n_clusters)

    def fit(self, X):
        self.model.fit(X)
        self.labels = self.model.labels_
        self.cluster_centers = self.model.cluster_centers_

    def score(self, X):
        return silhouette_score(X, self.labels)

    def plot_clusters(self):
        plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=self.labels, cmap='viridis')
        plt.scatter(self.cluster_centers[:, 0], self.cluster_centers[:, 1], c='red', marker='x')
        plt.title('KMeans Clustering')
        plt.show()
    def plot_silhouette(self, X):
        score = silhouette_score(X, self.labels)
        plt.figure(figsize=(8, 6))
        sns.histplot(self.labels, kde=True)
        plt.title(f'Silhouette Plot - KMeans (Score: {score:.2f})')
        plt.show()

# Class for Agglomerative Clustering
class AgglomerativeClusteringModel:
    def __init__(self, n_clusters=3):
        self.n_clusters = n_clusters
        self.model = AgglomerativeClustering(n_clusters=self.n_clusters)

    def fit(self, X):
        self.labels = self.model.fit_predict(X)

    def score(self, X):
        return silhouette_score(X, self.labels)

    def plot_clusters(self):
        plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=self.labels, cmap='viridis')
        plt.title('Agglomerative Clustering')
        plt.show()
    def plot_silhouette(self, X):
        score = silhouette_score(X, self.labels)
        plt.figure(figsize=(8, 6))
        sns.histplot(self.labels, kde=True)
        plt.title(f'Silhouette Plot - Agglomerative Clustering (Score: {score:.2f})')
        plt.show()

# Class for DBSCAN Clustering
class ClusteringWithPCA:
    def __init__(self, X_scaled, variance_threshold=0.95):
        self.X_scaled = X_scaled
        self.variance_threshold = variance_threshold
        self.pca = PCA(n_components=self.variance_threshold)  # PCA to preserve variance
        self.best_score = -1
        self.best_eps = 0
        self.best_min_samples = 0
        self.best_labels = None
        self.X_pca = self._apply_pca()

    def _apply_pca(self):
        # Apply PCA to the scaled data
        return self.pca.fit_transform(self.X_scaled)

    def perform_dbscan_grid_search(self, eps_range=None, min_samples_range=None):
        # Default ranges for epsilon and min_samples
        if eps_range is None:
            eps_range = np.arange(0.1, 2.1, 0.1)  # Default range for epsilon

        if min_samples_range is None:
            min_samples_range = range(2, 11)  # Default range for min_samples

        # Initialize lists to store scores for plotting
        silhouette_scores = []

        # Iterate over each combination of epsilon and min_samples
        for eps in eps_range:
            for min_samples in min_samples_range:
                # Initialize DBSCAN with the current epsilon and min_samples
                dbscan = DBSCAN(eps=eps, min_samples=min_samples)
                
                # Fit DBSCAN on the PCA-transformed data and predict cluster labels
                labels = dbscan.fit_predict(self.X_pca)

                # Calculate the number of clusters found (excluding noise points labeled as -1)
                n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
                
                # Proceed only if at least two clusters are found
                if n_clusters >= 2:
                    # Create a mask to exclude noise points from the evaluation
                    mask = labels != -1
                    
                    # Ensure there are enough points to calculate the silhouette score
                    if np.sum(mask) > 1:  # At least two non-noise points are required
                        # Calculate the silhouette score for the current clustering
                        score = silhouette_score(self.X_pca[mask], labels[mask])

                        # Update the best score and corresponding parameters if the current score is higher
                        silhouette_scores.append((eps, min_samples, score))
                        if score > self.best_score:
                            self.best_score = score
                            self.best_eps = eps
                            self.best_min_samples = min_samples
                            self.best_labels = labels

        return silhouette_scores

    def plot_scatter_graph(self, silhouette_scores):
        # Plot the scatter plot showing silhouette scores for various epsilon values
        eps_values = [score[0] for score in silhouette_scores]
        min_samples_values = [score[1] for score in silhouette_scores]
        scores = [score[2] for score in silhouette_scores]
        
        plt.figure(figsize=(10, 6))
        plt.scatter(eps_values, scores,c=min_samples_values, cmap='viridis')
        plt.colorbar(label="Min Samples")
        plt.xlabel('Epsilon')
        plt.ylabel('Silhouette Score')
        plt.title('Scatter plot for DBSCAN (Silhouette Score vs. Epsilon)')
        plt.show()

    def plot_clusters(self):
        # Scatter plot for the clusters based on the best DBSCAN result
        plt.figure(figsize=(10, 6))
        plt.scatter(self.X_pca[:, 0], self.X_pca[:, 1], c=self.best_labels, cmap='viridis')
        plt.title(f'DBSCAN Clusters (Epsilon={self.best_eps}, Min Samples={self.best_min_samples})')
        plt.xlabel('Principal Component 1')
        plt.ylabel('Principal Component 2')
        plt.colorbar(label='Cluster Label')
        plt.show()

    def display_best_results(self):
        if self.best_score > -1:
            print(f"Best parameters: eps={self.best_eps:.2f}, min_samples={self.best_min_samples}")
            print(f"Silhouette Score: {self.best_score:.4f}")
        else:
            print("No valid clustering solution found")

    def display_best_results(self):
        if self.best_score > -1:
            print(f"Best parameters: eps={self.best_eps:.2f}, min_samples={self.best_min_samples}")
            print(f"Number of clusters: {len(set(self.best_labels)) - (1 if -1 in self.best_labels else 0)}")
            print(f"Silhouette Score: {self.best_score:.4f}")
        else:
            print("No valid clustering solution found")
    def plot_silhouette(self, X):
        # Calculate silhouette scores for each sample
        silhouette_vals = silhouette_samples(X, self.best_labels)
    
        # Calculate the average silhouette score
        score = np.max(silhouette_vals)
        
        # Plot the silhouette values
        plt.figure(figsize=(8, 6))
        sns.histplot(silhouette_vals, kde=True, bins=30)
        plt.title(f'Silhouette Plot - DBSCAN (Avg Score: {score:.2f})')
        plt.show()              

def run_clustering_algorithms(X):

    X_scaled = StandardScaler().fit_transform(X)

    # KMeans
    kmeans = KMeans(n_clusters=3)
    kmeans.fit(X_scaled)
    kmeans_score = silhouette_score(X_scaled, kmeans.labels_)
    print("KMeans Silhouette Score: ", kmeans_score)

    # Agglomerative Clustering
    agglomerative = AgglomerativeClustering(n_clusters=3)
    agglomerative.fit(X_scaled)
    agglomerative_score = silhouette_score(X_scaled, agglomerative.labels_)
    print("Agglomerative Clustering Silhouette Score: ", agglomerative_score)
  
X = df
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
run_clustering_algorithms(X)
clustering = ClusteringWithPCA(X_scaled)

# Perform grid search for DBSCAN
silhouette_scores = clustering.perform_dbscan_grid_search()

# Plot the scatter plot to visualize how silhouette scores vary with epsilon and min_samples
clustering.plot_scatter_graph(silhouette_scores)

# Plot the clusters with the best DBSCAN result
clustering.plot_clusters()

# Display the best results
clustering.display_best_results()
############################## ELBOW PLOTS#####################################

# Elbow Plot for KMeans
def plot_kmeans_elbow(X_scaled, max_clusters=10):
    inertia = []
    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X_scaled)
        inertia.append(kmeans.inertia_)
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, max_clusters + 1), inertia, marker='o')
    plt.title('Elbow Plot for KMeans')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.show()

# Elbow Plot for Agglomerative Clustering
def plot_agglomerative_elbow(X_scaled, max_clusters=10):
    linkage_distances = []
    for k in range(2, max_clusters + 1):
        agglomerative = AgglomerativeClustering(n_clusters=k, compute_distances=True)  # Set compute_distances=True
        agglomerative.fit(X_scaled)
        # Calculate the linkage distance
        linkage_distances.append(agglomerative.distances_[-1])  # The last value of the linkage matrix
    plt.figure(figsize=(8, 6))
    plt.plot(range(2, max_clusters + 1), linkage_distances, marker='o')
    plt.title('Elbow Plot for Agglomerative Clustering')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Linkage Distance')
    plt.show()

# Elbow Plot for DBSCAN (Using silhouette score to evaluate clusters)
def plot_dbscan_elbow(X_scaled, eps_range=np.arange(0.1, 2.1, 0.1), min_samples_range=range(2, 11)):
    best_scores = []
    for eps in eps_range:
        for min_samples in min_samples_range:
            dbscan = DBSCAN(eps=eps, min_samples=min_samples)
            labels = dbscan.fit_predict(X_scaled)
            if len(set(labels)) > 1:  # Only valid if we have more than one cluster
                score = silhouette_score(X_scaled, labels)
                best_scores.append((eps, min_samples, score))
    
    # Find the best combination of eps and min_samples
    if best_scores:
        best_scores = sorted(best_scores, key=lambda x: x[2], reverse=False)
        best_eps, best_min_samples, best_score = best_scores[0]
        print(f"Best DBSCAN params: eps={best_eps}, min_samples={best_min_samples}, Silhouette Score={best_score:.2f}")
        plt.figure(figsize=(8, 6))
        plt.plot([x[2] for x in best_scores], [x[0] for x in best_scores], marker='o')
        plt.title('Elbow Plot for DBSCAN')
        plt.xlabel('Silhouette Score')
        plt.ylabel('Epsilon (eps)')
        plt.show()

# Plot Elbow for KMeans, Agglomerative, and DBSCAN
plot_kmeans_elbow(X_scaled)
plot_agglomerative_elbow(X_scaled)
plot_dbscan_elbow(X_scaled)
###############################################################################

# Scatter Plot for KMeans
def plot_kmeans_clusters(X_scaled, kmeans_labels):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=kmeans_labels, palette='Set1', s=100)
    plt.title('KMeans Clustering - Scatter Plot')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend(title='Cluster')
    plt.show()

# Scatter Plot for Agglomerative Clustering
def plot_agglomerative_clusters(X_scaled, agglomerative_labels):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=agglomerative_labels, palette='Set2', s=100)
    plt.title('Agglomerative Clustering - Scatter Plot')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend(title='Cluster')
    plt.show()

# Scatter Plot for DBSCAN
def plot_dbscan_clusters(X_scaled, dbscan_labels):
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=dbscan_labels, palette='Set3', s=100)
    plt.title('DBSCAN Clustering - Scatter Plot')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend(title='Cluster')
    plt.show()

# Define the clustering algorithms and their labels

# KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_scaled)
kmeans_labels = kmeans.labels_

# Agglomerative Clustering
agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative_labels = agglomerative.fit_predict(X_scaled)

# DBSCAN Clustering (Using the best eps and min_samples found previously)
best_eps = 1  # example value
best_min_samples = 2  # example value
dbscan = DBSCAN(eps=best_eps, min_samples=best_min_samples)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Plotting the scatter plots for each clustering method
plot_kmeans_clusters(X_scaled, kmeans_labels)
plot_agglomerative_clusters(X_scaled, agglomerative_labels)
plot_dbscan_clusters(X_scaled, dbscan_labels)

# Define a class for different regression models
class RegressionModels:
    def __init__(self, X, y):
        self.X = X
        self.y = y
        self.X_with_intercept = np.c_[np.ones(X.shape[0]), X]  # Add intercept term (X with ones)
    
    # Linear Regression using Normal Equation
    def linear_regression(self):
        theta = np.linalg.inv(self.X_with_intercept.T @ self.X_with_intercept) @ self.X_with_intercept.T @ self.y
        y_pred = self.X_with_intercept @ theta
        return y_pred

    # Polynomial Regression using Normal Equation
    def polynomial_regression(self, degree=3):
        X_poly = np.hstack([self.X_with_intercept] + [self.X_with_intercept[:, 1:]**i for i in range(2, degree + 1)])
        theta_poly = np.linalg.inv(X_poly.T @ X_poly) @ X_poly.T @ self.y
        y_pred_poly = X_poly @ theta_poly
        return y_pred_poly
    
    # Exponential Regression using Log Transformation
    def exponential_regression(self):
        # Using different features for exponential regression
        X_exp = self.X[:, [0, 1, 3, 4]]  # Using only specific columns for the exponential model (you can adjust as needed)
        X_with_intercept_exp = np.c_[np.ones(X_exp.shape[0]), X_exp]  # Add intercept term
        
        log_y = np.log(self.y)  # Apply log transformation to the target variable

        # Solve the normal equation for exponential regression
        theta_exp = np.linalg.inv(X_with_intercept_exp.T @ X_with_intercept_exp) @ X_with_intercept_exp.T @ log_y
        a = np.exp(theta_exp[0])  # Exponentiate to get 'a'
        b = theta_exp[1:]  # The coefficients for the exponential terms
        
        # Predictions for Exponential Regression
        y_pred_exp = a * np.exp(X_with_intercept_exp[:, 1:] @ b)  # Apply the exponential function for predictions
        return y_pred_exp

    # Method to calculate evaluation metrics (MSE, RMSE, R2)
    def metrics_cal(self, y_true, y_pred):
        mse = np.mean((y_true - y_pred) ** 2)
        rmse = np.sqrt(mse)
        r2 = 1 - (np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2))
        return mse, rmse, r2

    # Method to plot all models in one graph
    def plot_results(self, y_pred_linear, y_pred_poly, y_pred_exp):
        plt.figure(figsize=(10, 8))

        # Plot actual vs predicted for all models
        plt.scatter(self.y, y_pred_linear, color='blue', label='Linear Prediction', alpha=0.6)
        plt.scatter(self.y, y_pred_poly, color='orange', label='Polynomial Prediction', alpha=0.6)
        plt.scatter(self.y, y_pred_exp, color='green', label='Exponential Prediction', alpha=0.6)

        # Plotting the ideal line (y = x)
        plt.plot([self.y.min(), self.y.max()], [self.y.min(), self.y.max()], color='red', linestyle='--', label='Ideal Line (y=x)')

        # Labels and Title
        plt.xlabel("Actual Quality (Proline)")
        plt.ylabel("Predicted Quality (Proline)")
        plt.title("All Regression Models: Actual vs Predicted")
        plt.legend(loc='best')

        # Show the plot
        plt.show()

# Select relevant features for classification (example: Alcohol, Malic_Acid, etc.)
X = df[['Alcohol', 'Malic_Acid', 'Magnesium', 'Total_Phenols', 'Flavanoids', 'Proanthocyanins']].values
y = df['Proline'].values  # Assuming 'Proline' is used as the target variable for quality prediction

# Initialize the RegressionModels class
regression_models = RegressionModels(X, y)

# Perform Linear Regression
y_pred_linear = regression_models.linear_regression()

# Perform Polynomial Regression (degree 3)
y_pred_poly = regression_models.polynomial_regression(degree=3)

# Perform Exponential Regression
y_pred_exp = regression_models.exponential_regression()

# Calculate Metrics for each model
mse_linear, rmse_linear, r2_linear = regression_models.metrics_cal(y, y_pred_linear)
mse_poly, rmse_poly, r2_poly = regression_models.metrics_cal(y, y_pred_poly)
mse_exp, rmse_exp, r2_exp = regression_models.metrics_cal(y, y_pred_exp)

# Create a DataFrame to display the metrics
metrics_df = pd.DataFrame({
    'Model': ['Linear Regression', 'Polynomial Regression', 'Exponential Regression'],
    'MSE': [mse_linear, mse_poly, mse_exp],
    'RMSE': [rmse_linear, rmse_poly, rmse_exp],
    'R2 Score': [r2_linear, r2_poly, r2_exp]
})

print(metrics_df)

# Plot the results for all models
regression_models.plot_results(y_pred_linear, y_pred_poly, y_pred_exp)
```
